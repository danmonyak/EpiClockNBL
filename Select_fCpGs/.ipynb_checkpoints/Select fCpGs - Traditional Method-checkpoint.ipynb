```python
figsize=(8, 4)
sf=1

path_dict = {}
path_dict['tumor'] = os.path.join(proj_dir, 'cohort1.methyl.tsv')
# path_dict['normal'] = os.path.join(pc_consts['official_indir'], 'Normals', f'beta_values-{normals_TCGA_name}.tsv')

data = local_util.getDataDict(path_dict, filter_tum_samps=False, assert_same_sites=False)
data['tumor']['beta_values'] = data['tumor']['beta_values'].dropna(axis=1, how='all')
data['tumor']['pureSamples'] = data['tumor']['pureSamples'][np.isin(data['tumor']['pureSamples'], data['tumor']['beta_values'].columns)]
```

```python
outfile_path = os.path.join(proj_dir, 'beta_values_unbiased_sites.txt')

criteria = local_util.CLOCK_CRITERIA.copy()

# NO NORMAL SAMPLE
if 'normal' not in data['cohorts']:
    del criteria['normal']

# Return the Clock set of CpGs
neutral_DNA_CpG_list = local_util.getNeutralDNACpGs()
unbiased_sites = local_util.gen_CpG_set(data, criteria=criteria, neutral_DNA_CpG_list=neutral_DNA_CpG_list)
data['tumor']['beta_values_SELECTION'].loc[unbiased_sites].to_csv(outfile_path, sep='\t')
```

```python
# Define thresholds for unbiased CpGs
lower = 0.4
upper = 0.6

# Get beta values for all neutral CpGs
beta_values = data['tumor']['beta_means'].loc[neutral_DNA_CpG_list]

beta_values = beta_values.dropna()

# Create plot
fig, ax = plt.subplots(1, 1, figsize=figsize * sf)

# Create histogram and store the patches (bars)
bins = np.histogram_bin_edges(beta_values, bins=30)  # Define bin edges
n, _, patches = ax.hist(beta_values, bins=bins, color=pc_consts['palette_jco'][3],
                         alpha=pc_consts['opacity'], edgecolor="black", label="Biased CpGs")

# Manually color bins that fall within 0.4 and 0.6 range
for patch, bin_edge in zip(patches, bins[:-1]):  # Iterate over bins
    if lower <= bin_edge < upper - 0.01:
        patch.set_facecolor(pc_consts['palette_jco'][0])  # Change to blue
        patch.set_edgecolor("black")  # Keep black outline

# Customize figure
ax.set_xlabel('Average  in TARGET (Tumor)', fontsize=pc_consts['labelfontsize'] * sf)
ax.set_ylabel('Count', fontsize=pc_consts['labelfontsize'] * sf)
ax.set_title(f'Neutral CpGs (n = {len(neutral_DNA_CpG_list)})', fontsize=pc_consts['labelfontsize'] * sf)
ax.tick_params(axis='both', labelsize=pc_consts['ticksfontsize'] * sf, width=sf, length=8 * sf)

# Add legend manually
handles = [
    plt.Line2D([0], [0], color=pc_consts['palette_jco'][3], lw=4, label='Biased CpGs'),
    plt.Line2D([0], [0], color=pc_consts['palette_jco'][0], lw=4, label='Unbiased CpGs')
]
ax.legend(handles=handles, fontsize=pc_consts['labelfontsize'] * 0.8 * sf)

# Save figure
fig.savefig(os.path.join(figure_outdir, 'Figure_2_A_histogram.pdf'), format='pdf', pad_inches=0.1)
```

### Remove correlating sites

```python
# Remove sites that correlate with patient age

assert np.all((clinical['age_at_diagnosis'] + clinical['days_to_birth']).dropna() == 0)

spear_absCorr_list = []
patientID_list = beta_values_unbiased_sites.columns.map(pc_util.sampleToPatientID)
age_ser = clinical.loc[patientID_list, ['sampleID', 'age_at_diagnosis']].set_index('sampleID').squeeze('columns')
i = 0
for site in beta_values_unbiased_sites.index:
    if i % 10000 == 0:
        print(i)
    i += 1
    spear_absCorr_list.append(
        abs(
            spearmanr(
                pd.concat([beta_values_unbiased_sites.loc[site], age_ser], axis=1).dropna()
            ).correlation
        )
    )

spear_absCorr_ser = pd.Series(index=beta_values_unbiased_sites.index, data=spear_absCorr_list)
```

```python
sf = 1
num_sites = len(spear_absCorr_ser)

# Define threshold for coloring
threshold = 0.2

# Create plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the histogram and store the patches (bars)
bins = np.histogram_bin_edges(spear_absCorr_ser, bins=30)  # Define bin edges
n, _, patches = ax.hist(spear_absCorr_ser, bins=bins, edgecolor="black", alpha=0.75)

# Manually color bins: blue if  0.2, red if > 0.2
for patch, bin_edge in zip(patches, bins[:-1]):  
    if bin_edge <= threshold - 0.005:
        patch.set_facecolor(pc_consts['palette_jco'][0])
    else:
        patch.set_facecolor(pc_consts['palette_jco'][3])

# Customize labels and title
ax.set_xlabel(r'|Spearman $\rho$|', fontsize=pc_consts['labelfontsize'] * sf)
ax.set_title(f'Neutral and unbiased sites (centered mean in tumors)\n(n = {num_sites})',
             fontsize=pc_consts['labelfontsize'] * sf)

# Show plot
plt.show()

# Need to investigate this in other cancers
unbiased_uncorr_cpg_sites = spear_absCorr_ser[spear_absCorr_ser <= threshold]
print(unbiased_uncorr_cpg_sites.shape)
```

```python
## Figure 1C

# Plot inter-tumor standard deviation of beta for each unbiased site
# Color blue the Clock sites selected (most variable sites)

# Return the Clock set of CpGs
Clock_CpGs_final_set = local_util.gen_CpG_set(data, unbiased_uncorr_cpg_sites.index, n_select=1000, criteria=criteria)

print(len(Clock_CpGs_final_set))

# Create plot
fig, ax = plt.subplots(1, 1, figsize=figsize * sf)
sns.histplot(ax=ax,      # Non-fluctuating sites
             x=data['tumor']['beta_stds'].loc[unbiased_uncorr_cpg_sites.index[~np.isin(unbiased_uncorr_cpg_sites.index, Clock_CpGs_final_set)]],
             color=pc_consts['palette_jco'][3], alpha=pc_consts['opacity']
            )
sns.histplot(ax=ax,      # Fluctuating sites
             x=data['tumor']['beta_stds'].loc[Clock_CpGs_final_set],
             color=pc_consts['palette_jco'][0], alpha=pc_consts['opacity']
            )

# Customize figure
ax.set_xlabel('Intertumor standard deviation of  in TARGET', fontsize=pc_consts['labelfontsize'] * sf)
ax.set_ylabel(ax.get_ylabel(), fontsize=pc_consts['labelfontsize'] * sf)
ax.set_title(f'Neutral, unbiased, and uncorrelated CpGs (n = {len(unbiased_uncorr_cpg_sites)})', fontsize=pc_consts['labelfontsize'] * sf)
ax.tick_params(axis='both', labelsize=pc_consts['ticksfontsize'] * sf, width=sf, length=8 * sf)

# Label fluctuating sites
ax.text(0.235, 55, 'Fluctuating CpGs',
        ha="left", va="bottom", fontfamily='sans-serif', color='black',
        fontsize=pc_consts['labelfontsize'] * 0.8 * sf)

# Save figure
fig.savefig(os.path.join(figure_outdir, 'Figure_2_B.pdf'), format='pdf', pad_inches=0.1)
```

## Rest of selection process

```python
# Ideally, 1000 Clock sites selected
# We dont want to select more than half the number of unbiased sites (n)
# If n/2 < 1000, only pick n/2 sites
target_n = min(1000, int(balanced_CpGs.shape[0] / 2))
print(f'Picking {target_n} target sites')


########################
##### SELECT SITES #####
########################

# Fill missing values with mean of that site across tumors
Clock_CpGs_final_df = data['tumor']['beta_values_SELECTION'].loc[Clock_CpGs_final_set]
km_beta_values = Clock_CpGs_final_df.apply(lambda x:x.fillna(x.mean()), axis=1)

# Determine clustering weights
# See local_util for details
clustering_weights_list = [local_util.clusteringWeights(km_beta_values, random_state=0) for i in range(50)]
clustering_weights_ser = pd.concat(clustering_weights_list, axis=1).mean(axis=1)

# Calculate inter-tumor stdevs of sites
balanced_beta_stdevs = Clock_CpGs_final_df.std(axis=1).dropna()

# Pick the most variable sites
Clock_CpGs = balanced_beta_stdevs.sort_values(ascending=False).index.values[:target_n]

# Calculate c_beta using Clock sites
c_beta = 1 - Clock_CpGs_final_df.loc[Clock_CpGs].std(axis=0)

########################
###### SAVE FILES ######
########################

## Save Clock CpG list
Clock_CpGs_filepath = os.path.join(outdir, f'NBL_Clock_CpGs.txt')

# Save Clock sites, don't overwrite existing file
if os.path.exists(Clock_CpGs_filepath):
    existing_set = np.loadtxt(Clock_CpGs_filepath, dtype=str)
    assert np.all(np.sort(Clock_CpGs) == np.sort(existing_set))
    print('Current output matches existing file.')
else:
    np.savetxt(Clock_CpGs_filepath, Clock_CpGs, fmt='%s')

## Save c_beta values of final tumors
c_beta_final_samples = c_beta
c_beta_filepath = os.path.join(outdir, f'NBL.c_beta.txt')

# Save c_beta values, don't overwrite existing file
if os.path.exists(c_beta_filepath):
    existing_c_beta = pd.read_table(c_beta_filepath, index_col=0, header=None).squeeze("columns")
    assert (c_beta_final_samples - existing_c_beta < 1e-6).all()
    print('Current output matches existing file.')
else:
    c_beta_final_samples.to_csv(c_beta_filepath, sep='\t', header=False)

# Save beta values of Clock sites
outfile_path = os.path.join(outdir, f'NBL.methyl.traditional_sites.tsv')
Clock_CpGs_final_df.loc[Clock_CpGs].to_csv(outfile_path, sep='\t')
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

km_beta_values = beta_values_unbiased_sites.apply(lambda x:x.fillna(x.mean()), axis=1)
km = KMeans(n_clusters=4, random_state=None).fit(km_beta_values.T)
```

```python
def coloredPCA(col, use_groups=None, figsize=(6, 5), sites=None, ax=ax):
    if sites is None:
        sites = km_beta_values.index
    if ax is None:
        fig, ax = plt.subplots(figsize=figsize)
    
    tumors_transformed = PCA(2).fit_transform(km_beta_values.loc[sites].T)
    plot_data = pd.concat([pd.DataFrame(tumors_transformed, columns=['x', 'y']),
                       clinical.loc[km_beta_values.columns.map(pc_util.sampleToPatientID), col].reset_index()[col]
                      ], axis=1)
    if use_groups is not None:
        plot_data = plot_data.loc[plot_data[col].isin(use_groups)]
    sns.scatterplot(ax=ax, data=plot_data, x='x', y='y', hue=col)
```
